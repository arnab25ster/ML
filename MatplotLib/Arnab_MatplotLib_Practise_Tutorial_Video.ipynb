##### There are 2 ways of creating a subplot
- *Option 1*: We are creating the axis in the begining itself and then we are plotting against each axis.

```python
fig, ((ax1, ax2), (ax3, ax4))=plt.subplots(nrows=2,
                                          ncols=2,
                                          figsize=(10,7))
ax1.plot(x_axis, y_axis) #line plot
ax2.scatter(np.random.random(10), np.random.random(10)) # scatter plot
ax3.bar(nut_butter_price.keys(), height=nut_butter_price.values()) #barplot
ax4.hist(np.random.randn(100000)); #Horizontal bar plot
```

- *Option 2* : We are not creating the axis in the begining, but as we move on we are plotting against each axis giving the location of each axis as `ax[0,0]` and so on.

```python
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,7))
#
ax[0,0].plot(x_axis, y_axis),
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(nut_butter_price.keys(), height=nut_butter_price.values())
ax[1,1].hist(np.random.randn(100000));
```

#### Average Odometere reading based on KM's travelled by a car of particular Make
This is done by 2 examples 
- First in the old version of Pandas where we need to rename the column.
- In the second one the new version of Pandas where we dont have to rename the column. But only drawback with the newer  version of Panda is that w ecant have a column name with spaces in middle.

**Note:** in the below code we see the symbold as **`\`** in the lines... this is to show that the code of the line has not ended but continied in the next line. This avoids one single line of code to be long

**Note** The `reset-index()` function is also important. If we dont do it then while plotting we get a `Key Error`

```python
# Method 1 of renaming the columns explicitly

car_mean = car_sales.groupby('Make')\
.agg({"Make": "size", "Odometer (KM)": "mean"})\
.rename(columns={"Make":"Count of Cars", "Odometer (KM)": "Mean KM by Make"})
car_mean
```

```python
# method 2 of new version of Pandas
car_mean2 = car_sales.groupby("Make")\
.agg(Count_Cars = ("Make", "size"), Mean_Odometer = ("Odometer (KM)", "mean"))\
.reset_index()

car_mean2
```

```python
car_mean2.plot(x="Make", y="Mean_Odometer", kind="bar");
```

### Heart Disease dataset

```python
heart_disease = pd.read_csv(r"..\Input_Files\heart-disease.csv")
heart_disease.head()
```

```python
# create a histogram on the age columnn to see its distribution
heart_disease["age"].plot.hist(bins=45);
```

In the above graph we can see that there are few points which are sepearte from the main curve, they are called as **Outliers**.

The reason they are called as **Outliers** as they are more than 3 Standard deviation away from the mean.

```python
heart_disease.plot.hist(subplots=True, figsize=(10,50));
```

```python
over_age50 = heart_disease[heart_disease["age"] >= 50].reset_index(drop=True)
over_age50
```

```python
# Pyplot method
over_age50.plot(kind="scatter",
               x="age",
               y="chol",
               c="target");
```

**Note** in the above plot we dont have the labels mentioned as Axis

```python
# MIX way of doing, not purely OO method

fig, ax = plt.subplots(figsize=(10, 6))
over_age50.plot(kind="scatter",
               x="age",
               y="chol",
               c="target",
               ax=ax)

ax.set_xlim([45, 85]);  # This sets the limitations of X-axis 
```

```python
# OO method
fig, ax = plt.subplots(figsize=(10,6))

#Plot the data
scatter = ax.scatter(x=over_age50["age"],
                    y=over_age50["chol"],
                    c=over_age50["target"])

# Customize the plot
ax.set(title = "Heart Disease and Cholesterol Levels",
      xlabel = "Age",
      ylabel = "Cholesterol")

# Add a legend to the plot to determine the dots
ax.legend(*scatter.legend_elements(), title = "Target")

# Adding a mean value for the cholesterol levels
ax.axhline(over_age50["chol"].mean(),
          linestyle="--")
#ax.axhline(over_age50["chol"].std());
```

```python
over_age50["chol"].mean()
```

```python
over_age50["chol"].std()
```

```python
fig, (ax1, ax2) =  plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(8,10),
                               sharex=True)

# plotting the first axis
chol_scatter = ax1.scatter(x=over_age50["age"],
                          y=over_age50["chol"],
                          c=over_age50["target"])

ax1.set(title="Heart Disease and Cholesterol Levels",
      # xlabel="Age",  #as the X-axis is shared, no need to add it here.
       ylabel="Cholesterol")

ax1.legend(*chol_scatter.legend_elements(), title = "Heart Problems")

ax1.axhline(over_age50["chol"].mean(),
           linestyle="--")

# plotting the second axis
thalach_scatter = ax2.scatter(x=over_age50["age"],
                             y=over_age50["thalach"],
                             c=over_age50["target"])

ax2.set(title="Heart Disease and Heart rate",
       xlabel="Age",
       ylabel="Heart Rate")

ax2.legend(*thalach_scatter.legend_elements(), title = "Heart Problems")

ax2.axhline(over_age50["thalach"].mean(),
         linestyle="--");

fig.suptitle("Heart Disease Analysis", fontsize=15, fontweight="bold");
```

```python
# Set the style
plt.style.use("seaborn-whitegrid")

# OO method
fig, ax = plt.subplots(figsize=(10,6))

#Plot the data
scatter = ax.scatter(x=over_age50["age"],
                    y=over_age50["chol"],
                    c=over_age50["target"],
                    cmap="winter")

# Customize the plot
ax.set(title = "Heart Disease and Cholesterol Levels",
      xlabel = "Age",
      ylabel = "Cholesterol",
      xlim=[45, 80])    #customizing the X-axis limit from inside set

# Add a legend to the plot to determine the dots
ax.legend(*scatter.legend_elements(), title = "Target")

# Adding a mean value for the cholesterol levels
ax.axhline(over_age50["chol"].mean(),
           color=("red"),
           linestyle="--");
#ax.axhline(over_age50["chol"].std());
```

in the above code the X-axis limit was set from inside the set block by `xlim=[45, 80])` 
whereas for the below scatter plot the X-axis limit was set from outside the set block by `ax1.set_xlim([45, 80])`

**Note :-** when using it inside the set block we need to use the ` = ` symbol to assign where from outise it receives it as a function whihout the ` = ` symbol

```python
# Set the style
plt.style.use("seaborn-whitegrid")

fig, (ax1, ax2) =  plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(8,10),
                               sharex=True)

# plotting the first axis
chol_scatter = ax1.scatter(x=over_age50["age"],
                          y=over_age50["chol"],
                          c=over_age50["target"],
                          cmap="winter")

ax1.set(title="Heart Disease and Cholesterol Levels",
      # xlabel="Age",  #as the X-axis is shared, no need to add it here.
       ylabel="Cholesterol")

ax1.set_xlim([45, 80])

ax1.legend(*chol_scatter.legend_elements(), title = "Heart Problems")

ax1.axhline(over_age50["chol"].mean(),
            color="red",
           linestyle="--")

# plotting the second axis
thalach_scatter = ax2.scatter(x=over_age50["age"],
                             y=over_age50["thalach"],
                             c=over_age50["target"],
                             cmap="winter")

ax2.set(title="Heart Disease and Heart rate",
       xlabel="Age",
       ylabel="Heart Rate")

ax2.legend(*thalach_scatter.legend_elements(), title = "Heart Problems")

ax2.axhline(over_age50["thalach"].mean(),
            color="red",
         linestyle="--");

fig.suptitle("Heart Disease Analysis", fontsize=15, fontweight="bold");
```
